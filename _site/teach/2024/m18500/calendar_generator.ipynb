```python
import pandas as pd
import datetime

df = pd.read_csv('calendar.csv', index_col=0).fillna('')
df
```

```python
begin = datetime.datetime(2024, 3, 18)

df2 = df.copy()
df2.loc[:,:] = [['`' + (begin + datetime.timedelta(i * 7 + j)).strftime('%-m/%-d') + '` ' for j in range(5)] for i in range(10)]
(df2 + df).to_csv('calendar_with_date.csv')
```

```python
with open('calendar.md', 'w') as f:
    f.write((df2 + df).to_markdown())
print((df2 + df).to_markdown())
```
